# FM Demodulation of IQ Data -- Worked Example

This notebook uses about 0.4 seconds of raw I/Q data recorded by Yabi and Anthony from the RFSoC ADC at 64 MSps from the commercial FM band. We will walk through doing the DSP to isolate and demodulate an FM signal.

Start by loading the saved complex samples and plotting an FFT of a little bit of it:

```python
%matplotlib inline

import scipy
import numpy as np
import matplotlib.pyplot as plt

# Load data
c_data = np.load("raw_iq.npy")
sample_rate = 64e6
sample_period = 1.0 / sample_rate

# Take Fourier transform of 0.05 seconds of the data
n = int(0.05 / sample_period) # Number of samples in 0.05 seconds of data
fft = abs(np.fft.fftshift(np.fft.fft(c_data[0:n],n))) # abs() takes the magnitude. Magnitude of the Fourier-transformed data is the power at each frequency bin.

# Plot Fourier transform
frequency_bins = np.linspace(-sample_rate / 2.0, sample_rate / 2.0, n, endpoint = False) # We call the center 0 Hz. For 64 MSps, then our data goes from -32 MHz to +32 MHz
plt.plot(frequency_bins, fft)
plt.title("FFT of Raw RF Data")
plt.xlabel("RF Frequency")
plt.ylabel("Power")
plt.show()
```

We can see above that "0 Hz" is wherever the NCO of the RFSoC was set. I don't remember exactly what that was set to, somewhere in the middle of the commercial FM band like 80 MHz, probably. The big peaks are different broadcast FM radio stations.

We can see a big peak at 5.1658 MHz. This is a station. To mix it down to 0 Hz center frequency, we must generate a -5.1658 MHz local oscillator to mix it with. Because our local oscillator (NCO) is complex (it has an I and a Q component), it can have negative frequency. Functionally, negative frequency is actually just a different phase relationship between the I and Q oscillators.

```python
# Generate a complex NCO at -5.1658 MHz to mix the station at +5.1658 MHz down to zero IF
nco = -5.1658e6                                  # NCO frequency in Hz
omega_nco = 2 * np.pi * nco                      # NCO frequency in rad/sec
omega_discrete_nco = omega_nco / sample_rate     # NCO frequency discretized
sample_indices = np.arange(0, len(c_data))
complex_nco = np.e ** (1j * omega_discrete_nco * sample_indices)   # The real component of this will be the in-phase NCO, the imag will be the quadrature.

# Mix the signal with the complex NCO.
mixed = complex_nco * c_data
fft = abs(np.fft.fftshift(np.fft.fft(mixed[0:n],n))) # abs() takes the magnitude. Magnitude of the Fourier-transformed data is the power at each frequency bin.
plt.plot(frequency_bins, fft)
plt.title("FFT of Mixed-Down RF Data")
plt.xlabel("RF Frequency")
plt.ylabel("Power")
plt.show()
```

Note in the above spectrogram that we still have all the stations, we have just moved the desired station down to 0 Hz. We must now filter to select only the station we want. We _could_ do this by lowpassing I and Q, but the lowpass filters would have to have a huge number of taps since the fractional bandwidth of 200 kHz / 64 MHz is quite small. Using a CIC/decimation filter is easier:

```python
# CIC filter the mixed-down RF data (doing an FIR lowpass is impractical because it would have a huge number of taps)
decimation_factor = 320 # 64 MSps divided by 320 is 200 kSps or 200 kHz of bandwidth, which is the nominal channel width of broadcast FM
decimated_sample_rate = sample_rate / decimation_factor
filtered = scipy.signal.decimate(mixed, decimation_factor) # Not actually exactly a CIC filter, but does a similar thing. See scipy docs.
frequency_bins = np.linspace(-decimated_sample_rate / 2.0, decimated_sample_rate / 2.0, n, endpoint = False) # We call the center 0 Hz. For 64 MSps, then our data goes from -32 MHz to +32 MHz
fft = abs(np.fft.fftshift(np.fft.fft(filtered[0:n],n))) # abs() takes the magnitude. Magnitude of the Fourier-transformed data is the power at each frequency bin.
plt.plot(frequency_bins, fft)
plt.title("FFT of Mixed-Down and Filtered RF Data")
plt.xlabel("RF Frequency")
plt.ylabel("Power")
plt.show()
```

The big hump in the middle centered at 0 Hz is the analog FM signal. The two rectangular blocks are the "HD radio" digital signal, which uses a totally different modulation (OFDM). We need to get rid of these, as they will confuse our analog FM demodulation. There's not a good integer decimation factor that will do this, so let's do it using an FIR filter. Now that we're just trying to filter what looks to be about 55 kHz of the 200 kHz bandwidth, a small FIR filter can do it:

```python
fir_coeffs = scipy.signal.firwin(100, 55e3, fs = decimated_sample_rate)
fir_filtered = scipy.signal.lfilter(fir_coeffs, [1.0], filtered)

fft = abs(np.fft.fftshift(np.fft.fft(fir_filtered[0:n],n))) # abs() takes the magnitude. Magnitude of the Fourier-transformed data is the power at each frequency bin.
plt.plot(frequency_bins, fft)
plt.title("FFT of Mixed-Down and FIR Filtered RF Data")
plt.xlabel("RF Frequency")
plt.ylabel("Power")
plt.show()
```

Now at last we have just the analog FM signal. The spectrogram above doesn't convey much about the actual content of the signal because of how FM works. To see the actual content of the signal, we must demodulate by taking the angle of the discrete derivative of the signal:

```python
demodulated = 0.5 * np.angle(fir_filtered[0:-1] * np.conj(fir_filtered[1:])) # Frequency is the derivative of phase. We take the discrete derivative of the signal by multiplying with the time-delayed conjugate, then take angle.

frequency_bins = np.linspace(-decimated_sample_rate / 2.0, decimated_sample_rate / 2.0, len(demodulated), endpoint = False) # We call the center 0 Hz. For 64 MSps, then our data goes from -32 MHz to +32 MHz
fft = abs(np.fft.fftshift(np.fft.fft(demodulated,len(demodulated)))) # abs() takes the magnitude. Magnitude of the Fourier-transformed data is the power at each frequency bin.
plt.plot(frequency_bins/2, fft)
plt.title("FFT of Baseband Data")
plt.xlabel("Baseband Frequency")
plt.ylabel("Power")
plt.show()
```

Note that the demodulate signal is real, so positive and negative frequencies are the same. The x-axis is now baseband frequencies (frequencies of the modulating signal, not of the RF signal). We can see the 19 kHz pilot tone. I think we decimated and filtered too aggressively to see the 57 kHz RDS carrier. The audio seems to have been mostly removed too, maybe because our capture is short.
